```python
import shap
import joblib
import pandas as pd

# Load your test data (replace with your data)
X = pd.read_csv("../data/insurance_claims.csv")
model = joblib.load("../backend/eligibility_model.pkl")

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)

shap.summary_plot(shap_values, X, plot_type='bar')
```
